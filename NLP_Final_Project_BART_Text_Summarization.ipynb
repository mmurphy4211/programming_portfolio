```python
from transformers import pipeline
from datasets import load_dataset

# Load CNN/Daily Mail dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")

# Load a pre-trained summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Test the summarization pipeline on a sample article
sample_article = dataset["test"][0]["article"]  # Use the first article from the test set
summary = summarizer(sample_article, max_length=130, min_length=30, do_sample=False)

# Print the original article and summary
print("Original Article:\n", sample_article)
print("\nGenerated Summary:\n", summary[0]["summary_text"])
```
